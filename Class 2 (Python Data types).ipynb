# Python Data Types

# There are mainly four types of basic/primitive data types available in Python

1. Numeric: int, float, and complex
2. Sequence: String, list, and tuple
3. Set
4. Dictionary (dict)

# Numeric Data type

```python
# int - Integer

x=12
y=34

print(x)
print(y)
```

```python
# float

x=12.4
y=34.2

print(x)
print(y)
```

```python
# Compleax

x=12+3j
y=23+4j

print(x)
print(y)
```

# Sequence: String, list, and tuple

```python
# Str = String

x="Hello, My Name is Rahul"

print(x)
```

```python
y="1,2,3,4"

print(y)
```

```python
z="1+2+3"

print(z)
```

```python
# List  (used with square bracket)

a = ["apple", "banana", "cherry"]

print(a)
```

```python
b= ['Rahul',23,24.5,45+7j]

print(b)
```

```python
a+b
```

```python
# Tuple  (used with round bracket)

x = ("apple", "banana", "cherry")

print(x)
```

```python
y=("Rahul",45,56.3,65+8j)

print(y)
```

```python
x+y
```

# Set

```python
x1 = {"apple", "banana", "cherry"}

print(x1)
```

```python
x2={"Rahul",23,45.6,77+5j}

print(x2)
```

```python
# Frozenset

x3 = frozenset({"apple", "banana", "cherry"})
print(x3)
```

```python
x4 = frozenset({"apple", 43, "cherry",45.6,12+5j})
print(x4)
```
